# Bar Charts

## Basic Bar Chart

```scala
case class Species(name: String, amount: Int)
```

```scala
val bar = CustomPlotlyChart(Seq(Species("giraffes", 20), Species("orangutans", 14), Species("monkeys", 23)), 
                            dataOptions="{type: 'bar'}",
                            dataSources="{x: 'name', y: 'amount'}")
bar
```
